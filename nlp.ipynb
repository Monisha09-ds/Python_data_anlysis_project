```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
token = Tokenizer()
```

```python
text ="""NLP is a part of advanced Artificial Intelligence that teaches computers to understand human language. And what’s a better way to learn NLP than through projects? In this article, we will share the top 13 NLP projects that both beginners and experienced data professionals can use to understand better and work with language. These projects cover a wide range, from recognizing named entities to creating inspiring quotes. By working on these projects, you can use NLP to impact data analysis and processing."""
```

```python
token.fit_on_texts([text])
```

```python
token.word_index
```

```python
token.word_counts
```

```python
in_seq =[]
for sentence in text.split('\n'):
    #print(sentence)
    #print(token.texts_to_sequences([text]))
    tokenized_sen = token.texts_to_sequences([text])[0]
    #print(len(tokenized_sen))
    for i in range(1,len(tokenized_sen)):
       in_seq.append(tokenized_sen[:i+1])
    
    
```

```python
in_seq
```

```python
max_len =max([len(x) for x in in_seq])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
padded_inp = pad_sequences(in_seq,maxlen= max_len,padding ='pre')
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import KFold,cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
data = load_iris()
X,y = data.data,data.target
```

```python
K_f =KFold(n_splits=5,shuffle=True,random_state=42)
```

```python
model = LogisticRegression()
```

```python
scores = cross_val_score(model,X,y,cv=K_f,scoring='accuracy')
```

```python
print(f"Cross VAlidation Score :{scores}")
```

```python
print(f"Accuracy Score : {scores.mean()}")
```
